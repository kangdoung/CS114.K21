```
import pandas as pd
df = pd.read_csv("data_pre_v1.csv")
df.describe()
```

```
df.columns
```

```
y = df.Time
y.describe()
```

```
features = ['O(N)', 'O(N^2)', 'O(LogN)', 'O(NLogN)']
x = df[features]
x
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x , y)
model.score(x,y)
```

```
predict = model.predict(x)
predict
```

```
y
```

```
from sklearn.metrics import mean_absolute_error
mean_absolute_error(predict,y)
```
